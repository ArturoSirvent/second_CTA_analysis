```python
import tensorflow as tf 
import pickle
import os 
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
from glob import glob
```

```python
archivos=glob("../modelos/performances/*.pickle")
archivos
```

```python
for i in archivos:
    with open(i,"rb") as fil:
        hist_aux=pickle.load(fil)
    
    plt.figure(figsize=(13,13))
    plt.subplot(2,2,1)
    plt.plot(hist_aux.history["loss"],label="Loss")
    plt.plot(hist_aux.history["val_loss"],label="Loss val")
    plt.legend()
    plt.subplot(2,2,2)
    plt.plot(hist_aux.history["auc"],label="auc")
    plt.plot(hist_aux.history["val_auc"],label="auc val")
    plt.legend()
    plt.subplot(2,2,3)
    plt.plot(hist_aux.history["mean_squared_error"],label="mse")
    plt.plot(hist_aux.history["val_mean_squared_error"],label="mse val")
    plt.legend()
    plt.subplot(2,2,4)
    plt.plot(hist_aux.history["acc"],label="acc")
    plt.legend()
    

    

```
